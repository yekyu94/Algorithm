## List 생성

```python
class Node:
    def __init__(self, value = None, next = None):
        self.data = value
        self.next = next
    def __str__(self):
        return str(self.data)
```

```python
head = Node('a') # head
```

```python
n1 = Node('b')
head.next = n1
```

```python
n2 = Node('c')
n1.next = n2
```

```python
n3 = Node('d') # tail
n2.next = n3
```

```python
print(head, n1, n2, n3)
```

```python
print(head.next, n1.next, n2.next)
```

## 실제 구현

### Node 정의

```python
# Node 정의
class Node:
    def __init__(self, value = None, next = None):
        self.data = value
        self.next = next
    def __str__(self):
        return str(self.data)
```

### Node 추가

```python
# Node 추가 코드
def AppendNode(head, newData): # newData 새로운 노드에 들어갈 데이터
    newNode = Node(newData)
    if head.next == None:
        head.next = newNode
    else:
        currentNode = head.next
        while(currentNode.next != None):
            currentNode = currentNode.next
        currentNode.next = newNode
```

```python
head = Node('a')
AppendNode(head, 'b')
AppendNode(head, 'c')
AppendNode(head, 'd')
```

```python
print(head)
print(head.next)
print(head.next.next)
print(head.next.next.next)
```

### Node 삭제

```python
# Node 삭제 코드
def RemoveNode(head, remove): # remove는 노드가 아닌 노드가 갖고있는 값(value)
    if head.data == remove:
        head = head.next # 파이썬에서는 주소값문제로 이렇게 사용하면 head = RemoveNode(head, head.data) 로 받아야함
        return head
    elif head.next.data == remove:
        head.next = head.next.next
    else:
        currentNode = head.next
        while(currentNode.next != None and currentNode.next.data != remove):
            currentNode = currentNode.next
        if currentNode.next == None: return
        elif currentNode.next.data == remove:
            currentNode.next = currentNode.next.next
```

```python
head = RemoveNode(head, 'a') #head에 있는 값 제거시 "head =" 으로 받아야함
```

```python
RemoveNode(head, 'c')
```

```python
print(head)
print(head.next)
print(head.next.next)
```

### 노드 탐색 1

```python
# Node 탐색 코드
def getNodeValue(head, Location): # Location위치의 Node를 얻습니다.
    currentNode = head
    Location -= 1
    while(currentNode != None and Location >= 0):
        Location -= 1
        currentNode = currentNode.next
    return currentNode
```

```python
head = Node('a')
AppendNode(head, 'b')
AppendNode(head, 'c')
AppendNode(head, 'd')
print(head)
print(head.next)
print(head.next.next)
print(head.next.next.next)
```

```python
print(getNodeValue(head, 1))
```

### 노트 탐색 2

```python
def getNodeLoc(head, value):
    idx = 0
    if head.data == value:
        return idx
    else:
        currentNode = head.next
        idx += 1
        while(currentNode != None and currentNode.data != value):
            idx += 1
            currentNode = currentNode.next
        if currentNode == None: return '결과없음'
        else: return idx
```

```python
head = Node('a')
AppendNode(head, 'b')
AppendNode(head, 'c')
AppendNode(head, 'd')
print(head)
print(head.next)
print(head.next.next)
print(head.next.next.next)
```

```python
print(getNodeLoc(head, 'a'))
print(getNodeLoc(head, 'b'))
print(getNodeLoc(head, 'c'))
print(getNodeLoc(head, 'd'))
print(getNodeLoc(head, 'e'))
```
